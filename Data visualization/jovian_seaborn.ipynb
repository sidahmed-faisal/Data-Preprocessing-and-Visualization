### Plotting Multiple lines

```python
years = range(2000,2005)
apples=[0.23,0.33,0.25,0.34,0.54,]
oranges=[0.62,0.33,0.35,0.24,0.42]
```

```python
plt.plot(years , apples)
plt.plot(years , oranges)
plt.xlabel("Year")
plt.ylabel("Yields(tons per hectar)");
```

### Chart Title and legend

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel("year")
plt.xlabel("Yields(tons per hectare)")
plt.title("Crop Yields in kanto")
plt.legend(['Apples' , "Oranges"]);
```

### Line Markers

```python
plt.plot(years,apples ,marker = "X")
plt.plot(years,oranges,marker = "o")
plt.xlabel("year")
plt.xlabel("Yields(tons per hectare)")
plt.title("Crop Yields in kanto")
plt.legend(['Apples' , "Oranges"]);
```

```python
# Using fmt as a third argumant to specifiy color and marker and line
# fmt = [marker][line][color]
plt.plot(years,apples ,'s-b')
plt.plot(years,oranges,'o--r')
plt.xlabel("year")
plt.xlabel("Yields(tons per hectare)")
plt.title("Crop Yields in kanto")
plt.legend(['Apples' , "Oranges"]);
```

# without specifiying the line dots will be used instead

```python
plt.plot(years , oranges , 'or')
plt.title("yield of oranges");
```

### change the size of a figure

```python
plt.figure(figsize=(8,5))
plt.plot(years , oranges , 'or')
plt.title("yield of oranges");
```

 ## Scatter plot

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length , flowers_df.sepal_width)
```

```python
sns.set_style("darkgrid")
sns.scatterplot(flowers_df.sepal_length , flowers_df.sepal_width);
```

```python
sns.scatterplot(flowers_df.sepal_length ,
 flowers_df.sepal_width,
hue = flowers_df.species,
s =50);
```

### Mixing between matplotlib and seaborn

```python
plt.figure(figsize=(12,6))
plt.title("sepal flowres")
sns.scatterplot(flowers_df.sepal_length ,
 flowers_df.sepal_width,
hue = flowers_df.species,
s =50);
```

### Mixing between seaborn and pandas

```python
plt.figure(figsize=(12,6))
plt.title('sepal')
sns.scatterplot('sepal_length',
'sepal_width',

hue='species',
s = 100,
data = flowers_df)
```

```python
import numpy as np
```

```python
plt.figure(figsize=(10,6));
plt.title('Distribution of sapel width')
sns.histplot(flowers_df.sepal_width , bins=np.arange(2,5,0.25));
```

```python
plt.figure(figsize=(12,6));
plt.title('Distribution of sapel width')
sns.histplot(flowers_df.sepal_width , bins=[1,3,4,4.5]);
```

### Multiple Histograms

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df  = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
setosa_df.head()
```

```python
plt.figure(figsize=(12,6));
plt.hist(setosa_df.sepal_width , alpha=0.4 , bins = np.arange(2,5,0.25) , color = 'yellow')
plt.hist(versicolor_df.sepal_width  , alpha=0.4 , bins = np.arange(2,5,0.25) , color = 'r')
plt.legend(['Setosa' , 'Versicolor'])
```

```python
plt.figure(figsize=(12,6));
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width , versicolor_df.sepal_width 
        ,virginica_df.sepal_width],bins = np.arange(2,5,0.25),
        stacked=True);
plt.legend(['Setosa' , 'Versicolor' , 'Virginca'])
```

## Bar plots with averages

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
average =tips_df.groupby(['day' , 'size'])[['total_bill']].mean()
average
```

```python
average =tips_df.groupby('day')[['total_bill']].mean()
average
```

```python
plt.figure(figsize=(12,10))
average_sex =tips_df.groupby(['day','sex'])[['total_bill']].mean().plot(kind = "bar",legend= True ,stacked = True)
average_sex
```

```python
average_sex =tips_df.groupby(['day','sex'])[['total_bill']].mean().plot(kind = "pie",subplots=True,figsize=(15,12),legend= True ,stacked = True)
average_sex
```

```python
average_sex =tips_df.groupby(['day','sex'])[['total_bill']].mean().plot(kind = "bar",subplots=True,figsize=(15,12),legend= True ,stacked = True)
average_sex
```

```python
plt.figure(figsize=(12,6));

plt.bar(average.index , average['total_bill'])
```
