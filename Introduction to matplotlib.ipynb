## Matplotlib example workflow

```python
%matplotlib inline 
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [11,22,33,44]

fig , ax = plt.subplots(figsize = (10,10))
ax.plot(x,y);
plt.title("Simple plot");
plt.xlabel("x-axis");
plt.ylabel("y-axis");
```

## Making figures with Numpy arrays
we want: 
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
x = np.linspace(0,10,100)
x
```

```python
fig , ax = plt.subplots()
ax.plot(x , x**2);
```

```python
fig , ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig , ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
nut_butter = {"Almond butter":10,
              "peanut butter":8,
              "cashew butter":12
             }
```

```python
fig ,ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values());
plt.title("Butter store");
plt.ylabel("price $");
```

```python
fig , ax = plt.subplots()
ax.barh(list(nut_butter.keys()) , list(nut_butter.values()) );
```

```python
# Histogram 
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist(x);
```

## two options for subplots

```python
# Subplots option 1
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(x,x);
ax3.bar(nut_butter.keys(), nut_butter.values())
ax4.hist(np.random.random());
```

```python
# Subplots option 2
fig ,ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
```

## plotting from DataFrames

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
ts = pd.Series(np.random.randn(1000),
              index = pd.date_range("1/1/2020",periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"]= pd.date_range("1/1/2020",periods =len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# let's plot the total sales
car_sales.plot(x = "Sale Date" , y = "Total Sales");
```

```python
# converting price colomn to int from a string
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```
